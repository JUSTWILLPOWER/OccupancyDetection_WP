```python
import numpy as np 
import pandas as pd
from sklearn.metrics import zero_one_loss, mean_squared_error
import time
import matplotlib.pyplot as plt
import os
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib qt
```

```python
#get all required dataset files
tiVsAlgFiles = os.listdir('C:\\Users\\hasna\\Documents\\GitHub\\OccupancyDetection\\Results\\Algorithm vs TI')
groundTruthFiles = os.listdir('C:\\Users\\hasna\\Documents\\GitHub\\OccupancyDetection\\Results\\Ground Truth')
```

```python
#location datasets
#note X1 dataset - (x,y) = (-1.5,2.5)
#X2 dataset - (1.8,4)
#X3 dataset - (1.8,2.5)
```

### Misclassification and Accuracy

```python
missclassifications = np.array([])
accuracy = np.array([])
for name in tiVsAlgFiles:
    path = 'C:\\Users\\hasna\\Documents\\GitHub\\OccupancyDetection\\Results\\Algorithm vs TI\\' + name
    tiVsAlgDf = pd.read_csv(path, usecols=['TI', 'Group 16'], dtype=int)
    path = 'C:\\Users\\hasna\\Documents\\GitHub\\OccupancyDetection\\Results\\Ground Truth\\' + name
    groundTruthDf = pd.read_csv(path, usecols=['Expected Kalman'], dtype=int)
    maxNumberOfPeople = name[0]
    #remove rows with 111 in them (invalid rows)
    mask = np.logical_not(np.equal(groundTruthDf.values, 111))
    groundTruthDf = groundTruthDf[mask]
    tiVsAlgDf = tiVsAlgDf[mask]
    
    #compare the ground truth to TI
    if missclassifications.size == 0:
        missclassifications = np.array([zero_one_loss(groundTruthDf['Expected Kalman'], tiVsAlgDf['TI'])*100, 'TI', name,maxNumberOfPeople])
        accuracy = np.array([np.subtract(100,zero_one_loss(groundTruthDf['Expected Kalman'], tiVsAlgDf['TI'])*100), 'TI', name,maxNumberOfPeople])
        missclassifications = np.vstack((missclassifications,
                                        np.array([zero_one_loss(groundTruthDf['Expected Kalman'], tiVsAlgDf['Group 16'])*100, 'Group16',name,maxNumberOfPeople])))
        accuracy = np.vstack((accuracy, np.array([np.subtract(100,zero_one_loss(groundTruthDf['Expected Kalman'], tiVsAlgDf['Group 16'])*100), 'Group16',name,maxNumberOfPeople])))
    else:
        missclassifications = np.vstack((missclassifications,
                                        np.array([zero_one_loss(groundTruthDf['Expected Kalman'], tiVsAlgDf['TI'])*100, 'TI',name,maxNumberOfPeople])))
        accuracy = np.vstack((accuracy, np.array([np.subtract(100,zero_one_loss(groundTruthDf['Expected Kalman'], tiVsAlgDf['TI'])*100), 'TI',name,maxNumberOfPeople])))
        
        missclassifications = np.vstack((missclassifications,
                                        np.array([zero_one_loss(groundTruthDf['Expected Kalman'], tiVsAlgDf['Group 16'])*100, 'Group16',name,maxNumberOfPeople])))
    
        accuracy = np.vstack((accuracy, np.array([np.subtract(100,zero_one_loss(groundTruthDf['Expected Kalman'], tiVsAlgDf['Group 16'])*100), 'Group16',name,maxNumberOfPeople])))

missclassificationsDf = pd.DataFrame(missclassifications)
missclassificationsDf.columns = ['% Missclassifications', 'Group', 'Dataset', 'Number Of People']
missclassificationsDf['% Missclassifications'] = missclassificationsDf['% Missclassifications'].astype(dtype=np.float32)

accuracyDf = pd.DataFrame(accuracy)
accuracyDf.columns = ['Accuracy %', 'Group', 'Dataset', 'Number Of People']
accuracyDf['Accuracy %'] = accuracyDf['Accuracy %'].astype(dtype=np.float32)

fig, ax = plt.subplots()
fig.set_size_inches(11.8, 8.3)
missPlot = sns.barplot(y='Dataset', x='% Missclassifications', hue='Group', data=missclassificationsDf,orient='h')
missPlot.set_xticklabels(missPlot.get_xticklabels(), rotation=90)

fig, ax = plt.subplots()
fig.set_size_inches(11.8, 8.3)
accPlot = sns.barplot(y='Dataset', x='Accuracy %', hue='Group', data=accuracyDf,orient='h')
accPlot.set_xticklabels(accPlot.get_xticklabels(), rotation=90)
```

### Missclassification and Accuracy (Overall)

```python
#MISSCLASSIFICATIONS
#run code above first
#combine different people counts into overall amounts
missclassificationsDf['Number Of People'] = missclassificationsDf['Number Of People'].astype(dtype=int)
missclassificationOverall = np.array([])
for peopleCount in range(1,6):
    mask = np.equal(missclassificationsDf['Number Of People'],peopleCount)
    requiredFrame = missclassificationsDf[mask]
    tiMask = np.equal(requiredFrame['Group'], 'TI')
    missclassificationTI = np.mean(requiredFrame[tiMask]['% Missclassifications'])
    missclassificationG16 = np.mean(requiredFrame[np.logical_not(tiMask)]['% Missclassifications'])
    if missclassificationOverall.size == 0:
        missclassificationOverall = np.array([peopleCount,missclassificationTI, 'TI'])
        missclassificationOverall = np.vstack((missclassificationOverall,np.array([peopleCount,missclassificationG16, 'Group16'])))
    else:
        missclassificationOverall = np.vstack((missclassificationOverall,np.array([peopleCount,missclassificationTI, 'TI'])))
        missclassificationOverall = np.vstack((missclassificationOverall,np.array([peopleCount,missclassificationG16, 'Group16'])))
        
missclassificationOverallDf = pd.DataFrame(missclassificationOverall)
missclassificationOverallDf.columns = ['PeopleCount', 'Missclassification', 'Group']

fig, ax = plt.subplots()
fig.set_size_inches(11.8, 8.3)
missclassificationOverallDf['Missclassification']=missclassificationOverallDf['Missclassification'].astype(dtype=np.float32)
missOverallPlot = sns.barplot(x='PeopleCount',y='Missclassification',hue='Group', data=missclassificationOverallDf)

#ACCURACY
#run code above first
#combine different people counts into overall amounts
accuracyDf['Number Of People'] = accuracyDf['Number Of People'].astype(dtype=int)
accuracyOverall = np.array([])
for peopleCount in range(1,6):
    mask = np.equal(accuracyDf['Number Of People'],peopleCount)
    requiredFrame = accuracyDf[mask]
    tiMask = np.equal(requiredFrame['Group'], 'TI')
    accuracyTI = np.mean(requiredFrame[tiMask]['Accuracy %'])
    accuracyG16 = np.mean(requiredFrame[np.logical_not(tiMask)]['Accuracy %'])
    if accuracyOverall.size == 0:
        accuracyOverall = np.array([peopleCount,accuracyTI, 'TI'])
        accuracyOverall = np.vstack((accuracyOverall,np.array([peopleCount,accuracyG16, 'Group16'])))
    else:
        accuracyOverall = np.vstack((accuracyOverall,np.array([peopleCount,accuracyTI, 'TI'])))
        accuracyOverall = np.vstack((accuracyOverall,np.array([peopleCount,accuracyG16, 'Group16'])))
        
accuracyOverallDf = pd.DataFrame(accuracyOverall)
accuracyOverallDf.columns = ['PeopleCount', 'Accuracy', 'Group']

fig, ax = plt.subplots()
fig.set_size_inches(11.8, 8.3)
accuracyOverallDf['Accuracy']=accuracyOverallDf['Accuracy'].astype(dtype=np.float32)
accOverallPlot = sns.barplot(x='PeopleCount',y='Accuracy',hue='Group', data=accuracyOverallDf)
```

### RMSE (unit People)

```python
rmse = np.array([])
for name in tiVsAlgFiles:
    path = 'C:\\Users\\hasna\\Documents\\GitHub\\OccupancyDetection\\Results\\Algorithm vs TI\\' + name
    tiVsAlgDf = pd.read_csv(path, usecols=['TI', 'Group 16'], dtype=int)
    path = 'C:\\Users\\hasna\\Documents\\GitHub\\OccupancyDetection\\Results\\Ground Truth\\' + name
    groundTruthDf = pd.read_csv(path, usecols=['Expected Kalman'], dtype=int)
    
    #remove rows with 111 in them (invalid rows)
    mask = np.logical_not(np.equal(groundTruthDf.values, 111))
    groundTruthDf = groundTruthDf[mask]
    tiVsAlgDf = tiVsAlgDf[mask]
    
    if rmse.size == 0:
        rmseAlg = np.sqrt(mean_squared_error(groundTruthDf['Expected Kalman'], tiVsAlgDf['Group 16']))
        rmse = np.array([rmseAlg, 'Group16', name])
        rmseTi = np.sqrt(mean_squared_error(groundTruthDf['Expected Kalman'], tiVsAlgDf['TI']))
        rmse = np.vstack((rmse,np.array([rmseTi,'TI',name])))
    else:
        rmseAlg = np.sqrt(mean_squared_error(groundTruthDf['Expected Kalman'], tiVsAlgDf['Group 16']))
        rmse = np.vstack((rmse,np.array([rmseAlg, 'Group16',name])))
        rmseTi = np.sqrt(mean_squared_error(groundTruthDf['Expected Kalman'], tiVsAlgDf['TI']))
        rmse = np.vstack((rmse,np.array([rmseTi,'TI',name])))
    

rmseDf = pd.DataFrame(rmse)
rmseDf.columns = ['RMSE', 'Group', 'Dataset']
rmseDf['RMSE'] = rmseDf['RMSE'].astype(dtype=np.float32)

fig, ax = plt.subplots()
fig.set_size_inches(11.8, 8.3)
rmsePlot = sns.barplot(y='Dataset', x='RMSE', hue='Group', data=rmseDf, orient='h')    
```

## Location (KF Accuracy Graphs)

```python
xRMSE = np.array([])
yRMSE = np.array([])
for name in tiVsAlgFiles:
    try:
        path = 'C:\\Users\\hasna\\Documents\\GitHub\\OccupancyDetection\\Results\\Algorithm vs TI\\' + name
        tiVsAlgDf = pd.read_csv(path, usecols=['xTi', 'yTi', 'xGroup16', 'yGroup16'], dtype=np.float32)
        path = 'C:\\Users\\hasna\\Documents\\GitHub\\OccupancyDetection\\Results\\Ground Truth\\' + name
        groundTruthDf = pd.read_csv(path, usecols=['Expected X', 'Expected Y'], dtype=np.float32)
    except ValueError: #if not a location based dataset
        continue

    
    #remove rows with NaN in them (invalid rows)
    mask = np.logical_and(np.logical_not(np.isnan(tiVsAlgDf['xGroup16'])),np.logical_not(np.isnan(tiVsAlgDf['xTi'])))
    groundTruthDf = groundTruthDf[mask]
    tiVsAlgDf = tiVsAlgDf[mask]
    
    #calculate error in x direction (for both TI and Group16 Algorithm)
    if xRMSE.size == 0:
        rmseAlg = np.sqrt(mean_squared_error(groundTruthDf['Expected X'], tiVsAlgDf['xGroup16']))
        xRMSE = np.array([rmseAlg, 'Group16', name])
        rmseTi = np.sqrt(mean_squared_error(groundTruthDf['Expected X'], tiVsAlgDf['xTi']))
        xRMSE = np.vstack((xRMSE,np.array([rmseTi,'TI',name])))
    else:
        rmseAlg = np.sqrt(mean_squared_error(groundTruthDf['Expected X'], tiVsAlgDf['xGroup16']))
        xRMSE = np.vstack((xRMSE,np.array([rmseAlg, 'Group16',name])))
        rmseTi = np.sqrt(mean_squared_error(groundTruthDf['Expected X'], tiVsAlgDf['xTi']))
        xRMSE = np.vstack((xRMSE,np.array([rmseTi,'TI',name])))
        
    #calculate error in y direction (for both TI and Group16 Algorithm)
    if yRMSE.size == 0:
        rmseAlg = np.sqrt(mean_squared_error(groundTruthDf['Expected Y'], tiVsAlgDf['yGroup16']))
        yRMSE = np.array([rmseAlg, 'Group16', name])
        rmseTi = np.sqrt(mean_squared_error(groundTruthDf['Expected Y'], tiVsAlgDf['yTi']))
        yRMSE = np.vstack((yRMSE,np.array([rmseTi,'TI',name])))
    else:
        rmseAlg = np.sqrt(mean_squared_error(groundTruthDf['Expected Y'], tiVsAlgDf['yGroup16']))
        yRMSE = np.vstack((yRMSE,np.array([rmseAlg, 'Group16',name])))
        rmseTi = np.sqrt(mean_squared_error(groundTruthDf['Expected Y'], tiVsAlgDf['yTi']))
        yRMSE = np.vstack((yRMSE,np.array([rmseTi,'TI',name])))
        
    
rmse = np.hstack((np.expand_dims(yRMSE[:,0], axis=1),xRMSE))
rmseDf = pd.DataFrame(rmse)
rmseDf.columns = ['yRMSE','xRMSE', 'Group', 'Dataset']
rmseDf['xRMSE'] = rmseDf['xRMSE'].astype(dtype=np.float32)
rmseDf['yRMSE'] = rmseDf['yRMSE'].astype(dtype=np.float32)

#plot x y rmse
fig, ax = plt.subplots()
fig.set_size_inches(11.8, 8.3)
xrmsePlot = sns.barplot(x='Dataset', y='xRMSE', hue='Group', data=rmseDf)

fig, ax = plt.subplots()
fig.set_size_inches(11.8, 8.3)
yrmsePlot = sns.barplot(x='Dataset', y='yRMSE', hue='Group', data=rmseDf)
```
